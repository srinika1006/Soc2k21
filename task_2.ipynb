```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('master.csv')
```

```python
df.head()
```

```python
my_list=df.columns.tolist()
for col in my_list:
  df[col].hist()
  plt.show()
```

```python
my_list = ['suicides_no','population','suicides/100k pop']
for col in my_list:
  df[col]=pd.to_numeric(df[col])
  df.mode(axis=0,numeric_only=False,dropna=True).loc[0]
```
